```python
# sin カーブのグラフを作成

# 利用するライブラリをインポート
import matplotlib.pyplot as plt  # インポートしたライブラリを plt という名前で使う
import math

# データ用のリストとx軸の初期値を代入した変数とステップ幅を用意
x_value = 0.0 # x軸の初期値
x_step = 1.0 # forループの中で、ループを回すごとにx_valueに足す数
x = [] # グラフに表示する点のx座標
y = [] # グラフに表示する点のy座標 ※ xとyの点数は同じにする

# for ループを使ってデータを準備 (例) range(20) で20回ループを回す)
# x_valueをリストxに格納、sin(x_value)をリストyに格納し、x_valueにx_stepを足す
for i in range(10):
    x.append(x_value)
    y.append(math.sin(x_value))
    x_value = x_value + x_step

# print関数で、準備したデータのリストを表示します
print('x=', x)
print('y=', y)

# 準備したデータでグラフを作成します
plt.plot(x, y)
plt.show()
```
